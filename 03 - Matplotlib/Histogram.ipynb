# Part 4 : Plotting Histogram

```python
import matplotlib.pyplot as plt
from matplotlib import style
import random
import numpy as np
```

```python
ml_students_age = np.random.randint(18,45,100)
py_students_age = np.random.randint(15,45,100)
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.show()
```

```python
plt.hist(ml_students_age)
plt.show()
```

```python
plt.hist(ml_students_age, bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color='g',label = 'ML Student')

plt.title("ML Students Age Histogram")
plt.xlabel("Students age category")
plt.ylabel("No of student age")
plt.legend()
plt.show()
```

```python
plt.hist(ml_students_age, bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color='g',label = 'ML Student')
plt.hist(py_students_age, bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color='r',label = 'PY Student')

plt.title("ML Students Age Histogram")
plt.xlabel("Students age category")
plt.ylabel("No of student age")
plt.legend()
plt.show()
```

```python
# or above thing can be done by below code both are same things
plt.hist([ml_students_age,py_students_age], bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color=['g','r'],
         label = ['ML Student','PY Student'])

# plt.hist(py_students_age, bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color='r',label = 'PY Student')

plt.title("ML Students Age Histogram")
plt.xlabel("Students age category")
plt.ylabel("No of student age")
plt.legend()
plt.show()
```

```python
# or above thing can be done by below code both are same things

plt.figure(figsize=(16,9))

plt.hist([ml_students_age,py_students_age], bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color=['g','r'],
         label = ['ML Student','PY Student'])

# plt.hist(py_students_age, bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color='r',label = 'PY Student')

plt.title("ML Students Age Histogram")
plt.xlabel("Students age category")
plt.ylabel("No of student age")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))

plt.hist([ml_students_age,py_students_age], bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color=['g','r'],
         label = ['ML Student','PY Student'])

# plt.hist(py_students_age, bins = [15,20,25,30,35,40,45],rwidth=0.8,align='mid',color='r',label = 'PY Student')

plt.title("ML Students Age Histogram")
plt.xlabel("Students age category")
plt.ylabel("No of student age")
plt.legend()
plt.grid(color = 'k',linestyle = '--')
plt.show()
```
