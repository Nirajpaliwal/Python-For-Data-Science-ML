# Part 8 : Pie Chart

```python
from matplotlib import pyplot as plt
from matplotlib import style
import pandas as pd
import numpy as np
```

```python
classes = ["Python", "R", "Machine Learning", " Artificial Integgigance", "Data Science"]
clas1_students = [45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(clas1_students, labels = classes)
plt.show()
```

```python
explode = [0.03,0,0.1,0,0]
color = ['c','b','r','y','g']
textprops = {"fontsize" : 15}

plt.pie(clas1_students, labels = classes, explode = explode ,colors = color, autopct = "%0.1f%%",
       shadow = True, radius = 1.4, startangle = 270, textprops = textprops)

plt.show()
```

# Part 9 : Pie Chart Continue...

```python
explode = [0.03,0,0.1,0,0]
color = ['c','b','r','y','g']
textprops = {"fontsize" : 15}
wedgeprops = {"linewidth" : 2, "width" : 1, "edgecolor" : "k"}

plt.pie(clas1_students, 
        labels = classes, 
        explode = explode, 
        colors = color, 
        autopct = "%0.1f%%",
        shadow = True, 
        radius = 1, 
        startangle = 270, 
        textprops = textprops,
        pctdistance = 0.6,
        labeldistance = 1.2,
        counterclock = True, 
        wedgeprops = wedgeprops,
        center = (0,0),
        frame = True,
        rotatelabels = True)

plt.show()
```
