# Part 5 : Ploting Bar Chart

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ['Python', 'R', 'AI', 'ML', 'DS']
class1_students = [30,10,20,25,10] # out of 100 student in each class
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
plt.show()
```

```python
plt.barh(classes,class1_students)
plt.show()
```

```python
plt.bar(classes, class1_students, width=0.3, align='center', color = 'y', edgecolor = 'k',
       linewidth = 3, alpha = 0.8, linestyle = '--', label = 'Class1 Students') #visible = False

plt.legend()
plt.show()
```

# Part 6 : Plotting Bar Chart Continue...

```python
style.use('ggplot')
plt.figure(figsize=(12,6))

plt.bar(classes, class1_students, width=0.5, align='center', color = 'y', edgecolor = 'k',
       linewidth = 3, alpha = 0.8, linestyle = '--', label = 'Class1 Students' #visible = False
       )

plt.legend()
plt.grid(color='k',linestyle = '--')
plt.title('Bar Chart',fontsize = 15,fontweight = 1000)
plt.xlabel('Classes',fontsize = 13,fontweight = 1000)
plt.ylabel('No of student',fontsize = 13,fontweight = 1000)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(12,6))

plt.bar(classes, class1_students, width=0.3, color = 'g',label = 'Class1 Students') #visible = False
plt.bar(classes, class2_students, width=0.3, color = 'b',label = 'Class2 Students')

plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(12,6))

classes_index  = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class1_students, width, color = 'y', label = 'Class1 Students') #visible = False

plt.bar(classes_index + width, class2_students, width, color = 'r', label = 'Class2 Students')

plt.bar(classes_index + width + width, class3_students, width, color = 'g', label = 'Class3 Students')


plt.xticks(classes_index + width, classes, rotation = 50)


plt.legend()
plt.grid(color='k',linestyle = '--')
plt.title('Bar Chart',fontsize = 15,fontweight = 1000)
plt.xlabel('Classes',fontsize = 13,fontweight = 1000)
plt.ylabel('No of student',fontsize = 13,fontweight = 1000)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(12,6))

classes_index  = np.arange(len(classes))

width = 0.2

plt.barh(classes_index, class1_students, width, color = 'y', label = 'Class1 Students') #visible = False

plt.barh(classes_index + width, class2_students, width, color = 'r', label = 'Class2 Students')

plt.barh(classes_index + width + width, class3_students, width, color = 'g', label = 'Class3 Students')


plt.yticks(classes_index + width, classes, rotation = 50)


plt.legend()
plt.grid(color='k',linestyle = '--')
plt.title('Bar Chart',fontsize = 15,fontweight = 1000)
plt.ylabel('Classes',fontsize = 13,fontweight = 1000)
plt.xlabel('No of student',fontsize = 13,fontweight = 1000)
plt.show()
```
