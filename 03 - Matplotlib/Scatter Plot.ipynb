# Part 7 : Scatter Plot

```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
```

```python
df_google_play_store_apps = pd.read_csv('/Users/nirajpaliwal/Documents/Datasets/googleplaystore.csv', nrows = 1000)
```

```python
df_google_play_store_apps.head()
```

```python
df_google_play_store_apps.shape
```

```python
x = df_google_play_store_apps['Rating']
y = df_google_play_store_apps['Reviews']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize = (12,6))

plt.scatter(x,y)

plt.title('Google play store apps scatter plot',fontsize = 15,fontweight = 1000)
plt.xlabel('Rating',fontsize = 13,fontweight = 1000)
plt.ylabel('reviews',fontsize = 13,fontweight = 1000)
plt.show()
```

```python
plt.figure(figsize = (16,9))

plt.scatter(x,y, color = 'r',marker = 'o',s = 10,alpha = 0.8, linewidths=10, edgecolors='y')

plt.title('Google play store apps scatter plot',fontsize = 15,fontweight = 1000)
plt.xlabel('Rating',fontsize = 13,fontweight = 1000)
plt.ylabel('reviews',fontsize = 13,fontweight = 1000)
plt.show()
```

```python
plt.figure(figsize = (16,9))


plt.scatter(x,y, color = 'r',marker = '*',s = 100, alpha = 0.5, linewidth=10, edgecolor='g')
plt.scatter(x,df_google_play_store_apps['Installs'], color = 'y',marker = 'o',s = 100, alpha = 0.5, linewidths=10, edgecolors='c')


plt.title('Google play store apps scatter plot',fontsize = 15,fontweight = 1000)
plt.xlabel('Ratings')
plt.ylabel('Install')
plt.show()
```
