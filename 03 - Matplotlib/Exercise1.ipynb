```python
from matplotlib import pyplot as plt
# from matplotlib import style
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/nirajpaliwal/Documents/Python For Data Science & ML/Matplotlib/company_sales_data.csv')
```

```python
# Exercise 1: Read Total profit of all months and show it using a line plot
x = df['month_number'].tolist()
y = df['total_profit'].tolist()

# style.use('ggplot')

plt.plot(x,y,'bo--')

plt.title('Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])


# plt.grid(color='k', linestyle = '--')

plt.show()
```

```python
# Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

p = df['month_number'].tolist()
q = df['total_units'].tolist()

plt.plot(p,q,'ro--',label = 'profit data of last year',linewidth=3,markerfacecolor='k')
plt.legend(loc=4)
plt.title('company sales data of last yaer')
plt.xlabel('Month Number')
plt.ylabel('Total units')
plt.xticks(p)
plt.yticks([10000,20000,30000,40000,50000])
plt.show()
```

```python
# Exercise Question 3: Read all product sales data and show it  using a multiline plot

# plt.figure(figsize=(16,9))
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']
months = df['month_number'].tolist()

plt.plot(months,facecream,label='facecream sales data',color='r',marker='o',linewidth=4)
plt.plot(months,facewash,label='facewash sales data ',color='g',marker='o',linewidth=4)
plt.plot(months,toothpaste,label='toothpaste sales data',color='b',marker='o',linewidth=4)
plt.plot(months,bathingsoap,label='bathingsoap sales data',color='m',marker='o',linewidth=4)
plt.plot(months,shampoo,label='shampoo sales data',color='y',marker='o',linewidth=4)
plt.plot(months,moisturizer,label='moisturizer sales data',color='orange',marker='o',linewidth=4)

plt.title('sales data')
plt.xlabel('month number')
plt.ylabel('sales unt in numbers')
plt.xticks(months)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.legend()
plt.show()
```

```python
# Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot

plt.scatter(months,toothpaste,label='Toothepast sales data')
plt.grid(color='k',linestyle='--')
plt.title('toothepast sales data')
plt.xlabel('month number')
plt.ylabel('no of units sold')
plt.xticks(months)
plt.legend()
plt.show()
```

```python
# Exercise Question 5: Read face cream and facewash product sales data and show it using the bar chart
w = 0.2

m = np.arange(len(months))
month = df['month_number'].tolist()
plt.bar(m, facecream, width=w,label='Face cream sales data')
plt.bar(m+w, facewash, width=w, label = 'face wash sales data')

plt.title('facewash and facecream sales data')
plt.xlabel('month number')
plt.ylabel('sales units in number')
plt.legend(loc=2)
plt.xticks(m)
plt.grid(linestyle='--',linewidth=1)
plt.show()
```

```python
# Exercise Question 6: Read sales data of bathing soap of all months and show it 
# using a bar chart. Save this plot to your hard disk

plt.bar(months, bathingsoap)
plt.title('bathing soap sales data')
plt.xlabel('Month Numbers')
plt.ylabel('Sales units in number')

plt.grid(linestyle='--')
plt.xticks(month)
plt.show()
```

```python
# Exercise Question 7: Read the total profit of each month and show it using 
# the histogram to see most common profit ranges

total_profit = df['total_profit']
profit_range = [150000,175000,200000,225000,250000,300000,350000]

plt.hist(total_profit,profit_range,label='profit data')
plt.title('Profit Data')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual profit in dolar')
plt.legend(loc=2)
plt.show()
```

```python
# Exercise Question 8: Calculate total sale data 
# for last year for each product and show it using a Pie chart
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
products_count = np.arange(len(products))

plt.pie(products_count+1,labels=products,rotatelabels=True,autopct='%0.1f%%',shadow=True,
       labeldistance = 1.1,radius=1.2,startangle=10)
plt.title('Sales Data')
plt.show()
```
