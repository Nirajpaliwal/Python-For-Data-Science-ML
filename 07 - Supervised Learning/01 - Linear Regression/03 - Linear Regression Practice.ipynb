```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Employee_Salary.csv')
```

```python
# df = df.head(30)
```

```python
def func(data):
    return int(round(data,0))

def func2(data):
    return round(data,1)
```

```python
df['Salary'] = df['Salary'].apply(func)
```

```python
df['Years of Experience'] = df['Years of Experience'].apply(func2)
```

```python
df.head()
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
# X =  X.reshape(2000,1)
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
X_test
```

```python
y_prediction = lm.predict(X_test)
```
