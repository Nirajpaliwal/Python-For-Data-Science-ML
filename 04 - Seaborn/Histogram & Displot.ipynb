# Part 4 & 5 : Histogram / Displot

```python
from matplotlib import  pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
sns.distplot(tips_df['size'])

plt.show()
```

```python
sns.distplot(tips_df['tip'])

plt.show()
```

```python
sns.distplot(tips_df['total_bill'])

plt.show()
```

```python
bins = [0,10,20,30,40,50,60,70]
sns.distplot(tips_df['total_bill'], bins = bins)

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], hist=False)

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], hist=True,kde=False)

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], rug=True)

plt.show()
```

```python
# here we get kde and normalized line plot both
sns.distplot(tips_df['total_bill'], fit=norm)

plt.show()
```

```python
# here we get normalizes line plot
sns.distplot(tips_df['total_bill'], fit=norm, kde=False) 

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], color='r') 

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], vertical=True) 

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], norm_hist=True) 

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], axlabel='Total Billlll') 

plt.show()
```

```python
sns.distplot(tips_df['total_bill'], label='Total bill') 

plt.legend()
plt.show()
```

# Part 5 : Histogram / Displot Continue...

```python
sns.set(style='darkgrid')

sns.distplot(tips_df['total_bill'], label='Total bill') 

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))

bins = [1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df['total_bill'], bins = bins)

plt.xticks(bins)
plt.show()
```

```python
plt.figure(figsize=(16,9))


sns.distplot(tips_df['total_bill'], 
            hist_kws= {'color' : 'y','edgecolor':'k','linestyle':'--',
                      'linewidth': 2, 'alpha':0.9}) 

plt.show()
```

```python
plt.figure(figsize=(16,9))


sns.distplot(tips_df['total_bill'], 
            hist_kws= {'color' : 'y','edgecolor':'k','linestyle':'--',
                      'linewidth': 2, 'alpha':0.9},
            kde_kws = {'color' : 'k','linestyle':'--',
                      'linewidth': 2, 'alpha':0.9}) 

plt.show()
```

```python
plt.figure(figsize=(16,9))


sns.distplot(tips_df['total_bill'], 
            hist_kws= {'color' : 'y','edgecolor':'k','linestyle':'--',
                      'linewidth': 2, 'alpha':0.9},
            kde_kws = {'color' : 'k','linestyle':'--',
                      'linewidth': 2, 'alpha':0.9},
            rug=True,
            rug_kws={'color' : 'r','edgecolor':'b','linestyle':'--',
                      'linewidth': 1, 'alpha':0.9}) 

plt.show()
```

```python
plt.figure(figsize=(16,9))


sns.distplot(tips_df['total_bill'], 
            hist_kws= {'color' : 'y','edgecolor':'k','linestyle':'--',
                      'linewidth': 2, 'alpha':0.9},
            kde=False,
            fit=norm,
            fit_kws={'color' : 'b','linestyle':'--',
                      'linewidth': 3, 'alpha':0.9},
            rug=True,
            rug_kws={'color' : 'r','edgecolor':'b','linestyle':'--',
                      'linewidth': 1, 'alpha':0.9}) 

plt.show()
```
