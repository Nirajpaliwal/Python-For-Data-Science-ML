# Part 8 : Scatter Plot

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df.head()
```

```python
# titanic_df.shape
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex', style='who')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex', style='who', 
                size = 'who')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex', style='who', 
                size = 'who', sizes=(50,200))
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex', style='who', 
                size = 'who', sizes=(50,300))
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'who', y = 'fare', data = titanic_df, hue = 'alive', style='who', 
                size = 'alive', sizes=(50,300))
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x = 'who', y = 'fare', data = titanic_df, hue = 'alive', style='who', 
                size = 'alive', sizes=(50,300), palette="inferno_r", alpha=.7)
plt.show()
```
