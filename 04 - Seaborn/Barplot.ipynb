# Part 6 & 7 : Barplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df.head()
```

```python
sns.barplot()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex')

plt.show()
```

```python
order = ['Sun','Thur','Fri','Sat']
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex',order=order)

plt.show()
```

```python
hue_order = ['Female','Male']

sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', hue_order=hue_order)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', estimator=np.mean)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', estimator=np.max)

plt.show()
```

```python
# ci = confidance estimator
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', ci = 20)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', n_boot=3)

plt.show()
```

```python
sns.barplot(y="day", x='total_bill', data=tips_df, hue='sex')

plt.show()
```

```python
sns.barplot(x='total_bill', y='size', hue='sex', data=tips_df, orient='h')
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', color='b')

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', palette='hot')

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', saturation=.4)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', errcolor='r')

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', errwidth=8)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', capsize=.5)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, hue='sex', dodge=False)

plt.show()
```

# Part 7 : Barplot

```python
sns.set()
kwargs = {'alpha' : 0.8, 'linestyle' : '--', 'linewidth' : 5,
         'edgecolor' : 'k'}
sns.barplot(x="day", y='total_bill', data=tips_df, **kwargs)

plt.show()
```

```python
sns.barplot(x="day", y='total_bill', data=tips_df, alpha = 0.8, linestyle = '--',
           linewidth = 3, edgecolor = 'k')

plt.show()
```

```python
ax = sns.barplot(x="day", y='total_bill', data=tips_df, alpha = 0.8, linestyle = '--',
           linewidth = 3, edgecolor = 'k')

ax.set(title = "Barplot od Tips DataFrame",
        xlabel = "Days",
        ylabel = "Total Bill")

plt.show()
```

```python
plt.figure(figsize=(16,9))

sns.barplot(x="day", y='total_bill', data=tips_df, alpha = 0.8, linestyle = '--',
           linewidth = 3, edgecolor = 'k')


plt.title("Bar Plot Tips DF",fontsize = 20)
plt.xlabel("No of Days",fontsize = 16)
plt.ylabel("Total Bill",fontsize  =16)
# plt.savefig("Bar Plot of Days and Total Bill",dpi = 1000, quality=99)

plt.show()
```
