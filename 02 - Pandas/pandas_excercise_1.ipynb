```python
# Concatenate two data frames 
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
german_df = pd.DataFrame.from_dict(GermanCars)
german_df
```

```python
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
japanese_df = pd.DataFrame.from_dict(japaneseCars)
japanese_df
```

```python
pd.concat([german_df,japanese_df],keys=['German','Japanese'])
```

```python
# Merge two data frames
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}

price_df = pd.DataFrame.from_dict(Car_Price)
horsepower_df = pd.DataFrame.from_dict(car_Horsepower)
```

```python
price_df
```

```python
horsepower_df
```

```python
pd.merge(price_df,horsepower_df)
```
